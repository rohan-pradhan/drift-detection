```python
import matplotlib.pyplot as plt

def covar_shift_tool(train_data, prod_data):
    stacked_train_data = np.stack(train_data) 
    stacked_prod_data = np.stack(prod_data)
    
    avg_phi_val = []
    
    for x in range(0, 100): 
        train_samples, prod_samples, number_of_samples = sample_datasets(train_data, prod_data)
        avg_phi_val.append(log_reg(train_samples, prod_samples, number_of_samples))
        
    mean_phi = np.mean(avg_phi_val)
    
    
    
    if (mean_phi > 0.2):
        p_vals_index = []
        ks_vals, p_vals = ks_test(train_data,prod_data)
        for index, value in enumerate(p_vals):
#             print(x)
            if (value < 0.12):
                print(index)
                p_vals_index.append(index)
            
        p_vals_index.sort()
        
        
#         print (p_vals_index)
        
        ordinal = lambda n: "%d%s" % (n,"tsnrhtdd"[(n/10%10!=1)*(n%10<4)*n%10::4])
        print("We believe these features are drifting: ")
        plot_number = 1
        for x in p_vals_index:
            print(ordinal(x), " feature")
            plot_kde(train_data[x], prod_data[x], plot_number)
            plot_number = plot_number+1
            
#             s = pd.Series(train_x_vec[x])
#             ax = s.plot.kde("silverman")
#             s2 = pd.Series(prod_x_vec[x])
#             ax2 = s2.plot.kde("silverman")
        
    
        
    
    
    
def sample_datasets(train_data, prod_data):
    
    max_dataset_size = max(train_data.shape[0], prod_data.shape[0])

    number_of_samples = min(500, int(max_dataset_size*0.5)) 
    
#     print(number_of_samples)
    
    indices_train = np.random.randint(0,train_data.shape[0],number_of_samples)
    indices_prod = np.random.randint(0,prod_data.shape[0], number_of_samples)

    
    train_data = train_data[indices_train]
    
    prod_data = prod_data[indices_prod]
    
    return train_data, prod_data, number_of_samples


def log_reg(train_samples, prod_samples, number_of_samples):
    train_label = [1]*number_of_samples
    prod_label = [0]*number_of_samples
    
    full_data = np.concatenate([train_samples,prod_samples])
    full_labels = np.concatenate([train_label, prod_label])
    
    X_train, X_test, y_train, y_test = train_test_split(
    full_data, full_labels, test_size=0.2, random_state=42)


    logreg = LogisticRegression(solver="liblinear")
    logreg.fit(X_train, y_train)
    
    
    y_pred = logreg.predict(X_test)
    return (sklearn.metrics.matthews_corrcoef(y_test,y_pred))




def ks_test(train_data, prod_data):
    train_samples, prod_samples, number_of_samples = sample_datasets(train_data, prod_data)
    
    train_samples_transpose, prod_samples_transpose = train_samples.T, prod_samples.T
    
    ks_hash = []
    p_vals = []
    
    train_x_vec = []
    prod_x_vec = []
    
    
    for x in range(0, train_samples_transpose.shape[0]):
        train_data, prod_data  = train_samples_transpose[x], prod_samples_transpose[x]
#         train_x_vec.append(train_data)
#         prod_x_vec.append(prod_x_vec)
        ks_val = sp.stats.ks_2samp(train_data, prod_data)
        ks_hash.append(ks_val)
        p_vals.append(ks_val.pvalue)
        
    return ks_hash, p_vals
    
        
def plot_kde(train_data_column, prod_data_column, plot_number):
    plt.figure(plot_number)
    s = pd.Series(train_data_column)
    ax = s.plot.kde("silverman")
    s2 = pd.Series(prod_data_column)
    ax2 = s2.plot.kde("silverman")
    
    
    
covar_shift_tool(stacked_train_arrays, stacked_production_arrays)
    
    

# train_sample, production_sample = sample_datasets(stacked_train_arrays, stacked_production_arrays)

    
```
